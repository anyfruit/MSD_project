```python
# Define custom rules for each feature
custom_rules = {
    # C columns
    "C2_1": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_2": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_3": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_4": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_5": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_6": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_7": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_8": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C2_9": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C3": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    "C4": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3,
    # D2 columns
    "D2a_1": lambda x: 0 if x == 0 else 1 if x == 1 else 2 if x == 2 else 3,
    "D2a_2": lambda x: 0 if x == 0 else 1 if x == 1 else 2 if x == 2 else 3,
    # D4 and D6 columns
    "D4_1": lambda x: 0 if x <= 6 else 1 if 7 <= x <= 8 else 2,
    "D4_2": lambda x: 0 if x <= 6 else 1 if 7 <= x <= 8 else 2,
    "D6c": lambda x: 0 if x <= 6 else 1 if 7 <= x <= 8 else 2,
    # D7 column
    "D7c": lambda x: 0 if x <= 6 else 1 if 7 <= x <= 8 else 2,
    # F1a columns
    "F1a_1": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_2": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_3": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_4": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_5": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_6": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_7": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_8": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_9": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_10": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    "F1a_11": lambda x: 0 if x == 0 else 1 if x == 1 else 2,
    # F1b columns
    "F1b1": lambda x: 0 if x <= 6 else 1 if 7 <= x <= 8 else 2,
    "F1b2": lambda x: 0 if x <= 6 else 1 if 7 <= x <= 8 else 2,
    # G3 column
    "G3": lambda x: 0 if x <= 4 else 1 if x == 5 else 2 if x == 6 else 3
}

for feature, rule in custom_rules.items():
    if feature in df.columns:
        
        df[feature] = pd.to_numeric(df[feature], errors="coerce")
        
        df[f"{feature}_encoded"] = df[feature].apply(lambda x: rule(x) if pd.notnull(x) else None)
```
