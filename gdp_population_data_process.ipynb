```python
import pandas as pd
import numpy as np
```

```python
gdp_province = pd.read_excel("GDP_raw.xlsx", sheet_name="省市")
gdp_province.head()
```

```python
gdp_cn = pd.read_excel("GDP_raw.xlsx", sheet_name="全国")
gdp_cn = gdp_cn.iloc[:1,]
gdp_cn.head()
```

```python
gdp_years = []
for col in gdp_province.columns:
    if col.endswith("第四季度") or col == "2024年第三季度" or col == "地区":
        gdp_years.append(col)

gdp_proc = gdp_province[gdp_years]
gdp_proc.columns = [col[:4] for col in gdp_proc.columns]
gdp_proc.head()
```

```python
gdp_years = []
for col in gdp_cn.columns:
    if col.endswith("第四季度") or col == "2024年第三季度" or col == "地区":
        gdp_years.append(col)

gdp_cn = gdp_cn[gdp_years]
gdp_cn.columns = [col[:4] for col in gdp_cn.columns]
gdp_cn.head()
```

```python
gdp = pd.concat([gdp_proc,gdp_cn], ignore_index=True)
gdp.head()
```

```python
gdp_melted = gdp.melt(
    id_vars=["地区"],
    var_name="year",
    value_name="GDP"
)
gdp_melted.head()
```

```python
population_raw = pd.read_excel("population_raw.xlsx")

population_years = []
for col in population_raw.columns:
    if col.endswith("年") or col == "地区":
        population_years.append(col)

population_proc = population_raw[population_years]
population_proc.columns = [col[:4] for col in population_proc.columns]
population_proc.head()
```

```python
population_melted = population_proc.melt(
    id_vars=["地区"],
    var_name="year",
    value_name="总人口"
)
population_melted.head()
```

```python
final_df = pd.merge(population_melted, gdp_melted, on=["地区", "year"], how="right")
final_df["GDP"] = final_df["GDP"] * 100000000
final_df["总人口"] = final_df["总人口"] * 10000
```

```python
final_df.to_excel("gdp_population_data.xlsx", index=False)
```
